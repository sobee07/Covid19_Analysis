```python
#data colection for 2018
df=pd.DataFrame()
df_all=pd.DataFrame(columns=['type', 'geometry.type', 'geometry.coordinates', 'properties.timestamp',
       'properties.taxi_count', 'properties.api_info.status'])
list_date=pd.date_range(start='2018-01-01', end='2019-01-01', freq='1H')
for i in list_date:
  format_date=i.strftime('%Y-%m-%dT%H:%M:%S')
  payload = {'date_time':format_date}
  response = requests.get("https://api.data.gov.sg/v1/transport/taxi-availability",params=payload)
  json_response = response.json()
  df = pd.json_normalize(json_response['features'])
  df_all=df_all.append(df)
```

```python
# printing last date for 2018 extracted
print(max(df_all['properties.timestamp']))
```

```python
#Outputting the df to excel for year 2018
df_all.to_excel(r'Taxi_COunt_2018.xlsx', index = False)
```

```python
# Getting data for year 2020 till June 30
import requests
import pandas as pd
import numpy as np
import json
df=pd.DataFrame()
df_2020=pd.DataFrame(columns=['type', 'geometry.type', 'geometry.coordinates', 'properties.timestamp',
       'properties.taxi_count', 'properties.api_info.status'])
list_date=pd.date_range(start='2020-01-01', end='2020-06-30', freq='1H')
for i in list_date:
  format_date=i.strftime('%Y-%m-%dT%H:%M:%S')
  payload = {'date_time':format_date}
  response = requests.get("https://api.data.gov.sg/v1/transport/taxi-availability",params=payload)
  json_response = response.json()
  df = pd.json_normalize(json_response['features'])
  df_2020=df_2020.append(df)
```
